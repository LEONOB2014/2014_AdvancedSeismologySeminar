We can approximate an arbitrary function $f(x)$ using the interpolation with Lagrange polynomials $l_i$ at given collacation points $x_i$, i.e.

\begin{eqnarray*}
f(x) = \sum f(x_i) \cdot l_i(x).
\end{eqnarray*}

The Lagrange polynomials at $x$ are defined as follows:

$$ \ell_i^{(N)} (x) \ := \ \prod_{k = 1, \ k \neq i}^{N+1} \frac{x - x_k}{x_i-x_k}, \qquad   i = 1, 2, \dotsc , N + 1  $$

![Image](http://www.cb.uu.se/~cris/blog/matlabimages/lagrange_polynomials.png)


They are implemented in Python with the following code:

```python
def lagrange2(N,i,x, xi):

# Program to calculate Lagrange polynomials for order N
# and polynomial i [0, N] at location x for given collocation points x_i

   fac=1
   for j in range (-1,N): 
       if j != i:
          fac=fac*((x-xi[j+1])/(xi[i+1]-xi[j+1]))

       x=fac

   return x
```

```python
####---choose the exercise
exercise = 1


# Initialization
if exercise == 1:
    # Initialize space in the interval [-1 1]
    nx = 1000 # number of grid points
    x = np.linspace(-1, 1, nx)
    
    # Example function: sum over 5 sin function
   
    n = 5
    a = np.array([ -.5, -1, 3, 2, 5, -4]) 
    f = np.zeros(len(x))

################
# Please insert here your own function and change also the function fi
################

    for i in xrange(n):
        f = f + np.sin(np.pi / a[i] * x)

    
else: # exercice 2, Runge-function
    nx = 1000
    x = np.linspace(-5, 5, nx)

    #Runge function
    f = 1/(1 + x ** 2)

# Interpolate this function with Lagrange Polynomials of various degree,

# Get order of Lagrange polynomial
N = input(' Give polynomials degree (N): ')


if exercise==1:
    # Get collocation points xi from gll routine   
    [xi, w] = gll(N)
    xi = np.asarray(xi) # convert the list to an array
    
    # Initialize known function at collocation points
    fi = np.zeros(len(xi))
    
    for i in xrange(n):
       fi = fi + np.sin(np.pi / a[i] * xi) # calculate the function values at xi
    
        
else:
    xi=np.linspace(-5, 5, N+1)
    
    # Initialize known function at collocation points
    #Runge function
    fi= 1/(1+xi**2) # calculate the function values at xi

    
#Initialize Lagrange polynomials on the defined grid
lp = np.zeros((N+1,len(x)))
for i in xrange(0,len(x)):
    for j in xrange(-1,N):
            lp[j+1,i] = lagrange2(N,j,x[i],xi)

            
################################
# Calculate interpolating polynomials by multiplying 
# Lagrange polynomials with function values at xi
s = x*0
for j in xrange(0,N+1):   
    s = s + lp[j,:] * fi[j]

error = np.sum((np.abs(f-s))**2)/np.sum(np.abs(f))/100

# Plot results  
plt.figure()
plt.plot(x,s,'k-',color='green', label = 'Interpolating function')
plt.plot(x,f,'k--', label = 'Original function')
plt.plot(xi,fi,'s', label = 'Collocation points')
plt.title('Relative error: %g %%' % error)
plt.xlabel('x')
plt.ylabel('f(x)')


if exercise==1:
    plt.legend(loc = 'upper center')
else:
    plt.xlim(-5,5)
    plt.legend(loc = 2)
    


plt.show()

```
