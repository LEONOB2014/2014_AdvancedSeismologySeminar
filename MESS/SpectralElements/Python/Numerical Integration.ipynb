#### Fundamental principle for the numerical integration: 

Replace the function $f(x)$ that is going to be integrated by a polynomial approximation that can be integrated analytically.


As interpolating functions we use the Lagrange polynomials $l_i$ and obtain the following integration scheme for an arbitrary function $f(x)$ defined in the interval $[-1,1]$ :

\begin{eqnarray*} \int_{-1}^1 f(x) \ dx \approx \int _{-1}^1 P_N(x) dx = \sum_{i=1}^{N+1} w_i f(x_i) \end{eqnarray*}

with 
\begin{eqnarray*}
    P_N(x)= \sum_{i=1}^{N+1} f(x_i) \ l_i^{(N)}(x)
\end{eqnarray*}
As collacation points we use the Gauss-Lobatto-Legendre points $x_i$ and the corresponding weights that are needed to evaluate the integral are calculated as follows:

\begin{eqnarray*} w_i= \int_{-1}^1 l_i^{(N)}(x) \ dx \end{eqnarray*}.

```python
# This is a configuration step for the exercise. Please run it before the simulation code!
import numpy as np
import matplotlib.pyplot as plt

#import the gll routine to get x_i and w_i
from gll import *


# Show the plots in the Notebook.
plt.switch_backend("nbagg")
```

```python
# Exercise for Gauss integration


# Initialize space in [-1 1]
n = 100
x = np.linspace(-1,1,n)
f = x*x   # easy example: f=x^2
#f = np.exp(-x)

# analytical value of integral
intf = 1./3 * (1)**3 - 1./3 * (-1)**3     # The antiderivative of f = x^2 is [1/3 x^3] 
#intf = np.exp(-1)-np.exp(1)

# Calculate integral numerically
# Choose order
N = input('Give order of integration: ')

# Get integration points and weights from the gll routine
[xi,w] = gll(N)
xi=np.asarray(xi) #convert the list to an array

# Initialize function at points xi
fi = xi*xi
#fi = np.exp(-xi)

# Evaluate integral
intfn = 0
for i in range (0,len(w)-1):
    intfn = intfn + w[i]*fi[i]

#Plot results
error = np.abs(intfn-intf)/intf*100 
print 'Solution of the analytical integral: %g' %intf
print 'Solution of the numerical integral: %g' %intfn
print 'Error numerical/analytical solution %g %% ' %error


plt.plot(x,f,'k-', label='analytical')
plt.plot(xi,fi,'b--', label='numerical')
plt.plot(xi, fi,'bs', label='GLL points')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Numerical vs. analytical function')


#plt.title(' Analytical: %g Numerical: %g Error: %g ' %(intf,intfn, np.abs(intfn-intf)/intf*100))
plt.legend()#(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()

```
